```python
def month_to_number(month_name):
    months = {
        "janvier":"01",
        "fevrier":"02",
        "mars":"03",
        "avril":"04",
        "mai":"05",
        "juin":"06",
        "juillet":"07",
        "aout":"08",
        "septembre":"09",
        "octobre":"10",
        "novembre":"11",
        "decembre":"12"
    }
    return(months.get(month_name, "00"))
```
